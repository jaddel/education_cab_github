```python
phones = pd.read_csv(r'C:\Users\Vigi\CAB - work related\eCommerce\individual datasets\smartphone.csv')
phones.head(3)
```

```python
# remove $ sign
phones.price = phones.price.str.replace('$', '')

# %%
# keep only the value of the rating
phones.rating = phones.rating.str[:3]

# drop null values in title column
phones=phones.dropna(subset=['title'])
phones=phones.sort_values(by=['title'])
phones.head()
phones.isnull().any()
```

```python
# make sure that Price column is data type int
phones=phones.sort_values(by=['price'],ascending=False)
phones.price=phones.price.str.replace(',','').astype(float)
phones
#phones.price=phones.price.astype(float)
#phones.dtypes
```
